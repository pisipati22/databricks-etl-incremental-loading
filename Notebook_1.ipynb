#DATA WAREHOUSING

```sql
CREATE DATABASE salesDWH
```

#STAGING

```sql
CREATE OR REPLACE TABLE salesDWH.stg_sales
AS
SELECT * FROM sales.orders WHERE OrderDate>'2022-01-05'
```

#TRANSFORMATION

```sql
CREATE VIEW salesDWH.trans_sales AS
SELECT * FROM salesDWH.stg_sales WHERE Quantity IS NOT NULL
```

```sql
CREATE DATABASE orderDWH
```

```sql
CREATE OR REPLACE TABLE orderDWH.stg_sales
AS
SELECT * FROM sales_NEW.orders
```

```sql
CREATE VIEW orderDWH.trans_sales AS
SELECT * FROM orderDWH.stg_sales WHERE Quantity IS NOT NULL
```

```sql
SELECT * FROM orderDWH.trans_sales
```

```sql
CREATE OR REPLACE TABLE orderDWH.DimCustomers
(
  CustomerID INT,
  DimCustomerKey INT,
  CustomerName STRING,
  CustomerEmail STRING
)
```

```sql
CREATE OR REPLACE VIEW orderDWH.view_DimCustomers AS
SELECT 
ROW_NUMBER() OVER(ORDER BY CustomerID) AS DimCustomerKey,
CustomerID,
CustomerName,
CustomerEmail
FROM orderDWH.trans_sales
```

```sql
INSERT INTO orderDWH.Dimcustomers
SELECT * FROM orderDWH.view_DimCustomers
```

```sql
CREATE TABLE orderDWH.DimProduct
(
 ProductID INT,
  DimProductKey INT,
  ProductName STRING,
 ProductCategory STRING
)
```

```sql

CREATE OR REPLACE VIEW orderDWH.view_DimProduct AS
SELECT 
ROW_NUMBER() OVER(ORDER BY ProductID) AS DimProductKey,
ProductID,
ProductName,
ProductCategory
FROM orderDWH.trans_sales
```

```sql
INSERT INTO orderDWH.dimproduct
SELECT * FROM orderDWH.view_DimProduct
```

```sql
CREATE TABLE orderDWH.DimRegion
(
 RegionID INT,
  RegionName STRING,
  Country STRING,
 DimRegionKey INT
)
```

```sql
CREATE OR REPLACE VIEW orderDWH.view_DimRegion AS
SELECT T.*, 
ROW_NUMBER() OVER(ORDER BY RegionID) AS DimRegionKey FROM
(
SELECT 
DISTINCT(RegionID) AS RegionID,
RegionName,
Country
FROM orderDWH.trans_sales) AS T
```

```sql
INSERT INTO orderDWH.dimRegion
SELECT * FROM orderDWH.view_dimRegion
```

```sql
SELECT * FROM orderDWH.DimRegion
```

```sql
CREATE OR REPLACE TABLE orderDWH.DimDate
(
 OrderDate DATE,
 DimDateKey INT
)
```

```sql
CREATE OR REPLACE VIEW orderDWH.view_DimDate AS
SELECT T.*, 
ROW_NUMBER() OVER(ORDER BY T.OrderDate) AS DimDateKey FROM
(
SELECT 
DISTINCT(OrderDate) AS OrderDate
FROM orderDWH.trans_sales) AS T
```

```sql
INSERT INTO orderDWH.dimDate
SELECT * FROM orderDWH.view_DimDate
```

##FACT TABLE

```sql
CREATE TABLE orderDWH.FactSales(
OrderID INT,
Quantity INT,
UnitPrice DECIMAL,
Total DECIMAL,
DimProductKey INT,
DimCustomerKey INT,
DimRegionKey INT,
DimDateKey INT)
```

```sql
SELECT 
F.OrderID,
F.Quantity,
F.UnitPrice,
F.Total,
DC.DimCustomerKey,
DP.DimProductKey,
DR.DimRegionKey,
DD.DimDateKey FROM orderDWH.trans_sales F
LEFT JOIN orderDWH.DimCustomers DC
ON F.CustomerID = DC.CustomerID
LEFT JOIN orderDWH.DimProduct DP
ON F.ProductID = DP.ProductID
LEFT JOIN orderDWH.DimRegion DR
ON F.RegionID = DR.RegionID
LEFT JOIN orderDWH.DimDate DD
ON F.OrderDate = DD.OrderDate
```
